# Air Pollution Analysis using Python


### Air pollution is a mixture of solid particles and gases in the air. Car emissions, chemicals from factories, dust, pollen and mold spores may be suspended as particles.Some air pollutants are poisonous. Inhaling them can increase the chance you'll have health problems. People with heart or lung disease, older adults and children are at greater risk from air pollution. Air pollution isn't just outside - the air inside buildings can also be polluted and affect your health.

### Graph showing different pollutants .

```python
plt.figure(figsize=(12,8))
style.use("ggplot")
x=df["State"].to_numpy()
y=df["SO2 Annual Avg"].to_numpy()
y1=df["NO2 Annual Avg"].to_numpy()
y2=df["PM10 Annual Avg"].to_numpy()
y3=df["PM25 Annual Avg"].to_numpy()
plt.plot(x,y,"c",mec="b",label="SO2",linewidth=2)
plt.plot(x,y1,"g",mec="b",label="NO2",linewidth=2)
plt.plot(x,y2,"r",mec="b",label="PM10",linewidth=2)
plt.plot(x,y3,"orange",mec="b",label="PM25",linewidth=2)
plt.title("Literacy Rate of 2001 & 2011 in Urban Areas")
plt.ylabel("Literacy Rate")
plt.xlabel("State")
plt.legend()
plt.xticks(x,rotation="vertical")
plt.grid(True,color="w")
plt.show()
```

### Graph showing the yearwise record of NO2 pollutants in the Nation.

```python
plt.figure(figsize=(10,6))
y=ydf["NO2 Annual Avg"].to_numpy()
l=np.array([2016,2017,2018])
plt.pie(y,labels=l,shadow=True,startangle=90,explode=[0.1,0,0],autopct="%1.1f%%",wedgeprops={"edgecolor":"black"})
plt.show()
```
